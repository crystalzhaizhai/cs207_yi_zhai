# Problem 1:  Linked List Class
Write a linked list class called `LinkedList`.  Remember, a singly linked list is made up of nodes each of which contain a value and a pointer.  The first node is called the "head node".

Here are the required methods:
* `__init__(self, head)` where `head` is the value of the head node.  You could make the head node an attribute.
* `__len__(self)`: Returns the number of elements in the linked list.
* `__getitem__(self, index)` returns the value of the node corresponding to `index`.  Include checks to make sure that `index` is not out of range and that the user is not trying to index and empty list.
* `__repr__(self)` returns `LinkedList(head_node)`.
* `insert_front(self, element)` inserts a new node with value `element` at the beginning of the list.
* `insert_back(self, element)` inserts a new node with value `element` at the end of the list.

Note:  An alternative implementation is to create a `Node` class.  You are not required to make a `Node` class but you may if you prefer that implementation.  Please don't steal that implementation from the online forums.  I've seen those too.

````python
class Node:
    def __init__(self,value=0):
        self.value=value
        self.next=None
        
class LinkedList:
    def __init__(self, head):
        self.head=Node(head)
    def __len__(self):
        point=self.head
        count=0
        while (point.next!=None):
            point=point.next
            count+=1
        return count
    
    def __getitem__(self,index):
        point=self.head
        count=index
        while (point.next!=None) and count!=0:
            point=point.next
            count-=1
        if (count!=0):
            raise ValueEroor("index exceeding the length of the list")
        return point.value
    def __repr__(self):
        return "LinkedList(self.head(%d))"%self.head.value
    
    def insert_front(self,element):
        new_ele=Node(element)
        new_ele.next=self.head.next
        self.head.next=new_ele
    def insert_back(self,element):
        new_ele=Node(element)
        point=self.head
        while (point.next!=None):
            point=point.next
        point.next=new_ele
        
        
    
````

# Problem 2:  Binary Tree Class
A binary search tree is a binary tree with the invariant that for any particular node the left child is smaller and the right child is larger. Create the class `BinaryTree` with the following specifications:

`__init__(self)`: Constructor takes no additional arguments

`insert(self, val)`: This method will insert `val` into the tree

(Optional) `remove(self, val)`: This will remove `val` from the tree.
1. If the node to be deleted has no children then just remove it.
2. If the node to be deleted has only one child, remove the node and replace it with its child.
3. If the node to be deleted has two children, replace the node to be deleted with the maximum value in the left subtree.  Finally, delete the node with the maximum value in the left-subtree.

`getValues(self. depth)`: Return a list of the entire row of nodes at the specified depth with `None` at the index if there is no value in the tree. The length of the list should therefore be $2^{\text{depth}}$. 

````python
class treenode():
    def __init__(self,value):
        self.value=value
        self.left=None
        self.right=None
        self.father=None
    def __repr__(self):
        return str(self.value)
    
class BinaryTree:
    def __init__(self):
        self.root=None
    def insert(self,val):
        newnode=treenode(val)
        if self.root==None:
            self.root=newnode
            return
        point=self.root
        while 1!=0:
            if point.value>val:
                if point.left!=None:
                    point=point.left
                else:
                    point.left=newnode
                    newnode.father=point
                    return
            if point.value<val:
                if point.right!=None:
                    point=point.right
                else:
                    point.right=newnode
                    newnode.father=point
                    return
            if point.value==val:
                raise ValueError("inserted value already there")

    def remove(self,val):
        point=self.root
        
        while point!=None:
            if point.value==val:
                del_point=point
                break
            else:
                if point.value>val:
                    
                    if point.left!=None:
                        currentroad=0
                        point=point.left
                    else:
                        raise ValueError("value not found")
                else:
                    if point.right!=None:
                        currentroad=1
                        point=point.right
                    else:
                        raise ValueError("value not found")
        if del_point.left==None:
            if del_point.father!=None:
                if currentroad==0:
                    del_point.father.left=del_point.right
                    if del_point.right!=None:
                        del_point.right.father=del_point.father
                elif currentroad==1:
                    del_point.father.right=del_point.right
                    if del_point.right!=None:
                        del_point.right.father=del_point.father
            else:
                if currentroad==0:
                    self.root=del_point.right
                elif currentroad==1:
                    self.root=del_point.right
            
        else:
            point2=del_point.left
            if point2.right==None:
                if del_point.father!=None:
                    point2.father=del_point.father
                    point2.right=del_point.right
                    if currentroad==0:
                        del_point.father.left=point2
                    elif currentroad==1:
                        del_point.father.right=point2
                else:
                    self.root=point2
                    point2.right=del_point.right
                    point2.left=del_point.left
            else:
                while point2.right!=None:
                    point2=point2.right
               
                if point2.left!=None:
                    point2.father.right=point2.left
                    point2.left.father=point2.father
                if del_point.father!=None:
                    if currentroad==0:
                        del_point.father.left=point2
                        del_point.right.father=del_point.father
                    elif currentroad==1:
                        del_point.father.right=point2
                        del_point.right.father=del_point.father
                    point2.father=del_point.father
                    point2.right=del_point.right
                    point2.left=del_point.left
                else:
                    self.root=point2
                    point2.right=del_point.right
                    point2.left=del_point.left
                return
            
        
    def getValues(self, depth):
        point=self.root
        father_list=[self.root]
        child_list=[]
        for i in range(depth):
            for node in father_list:          
                if node!=None:
                    child_list.append(node.left)
                    child_list.append(node.right)
                else: 
                    child_list.append(None)
                    child_list.append(None)
            father_list=child_list.copy()
            child_list=[]
        return(father_list)
        
       
            
            
    def __len__(self):
        
        point=self.root
        if self.root!=None:
            father_list=[self.root]
        child_list=[]
        depth=0
        while len(father_list)!=0:
            depth+=1
            for node in father_list:
                if node!=None:
                    if node.left!=None:
                        child_list.append(node.left)
                    if node.right!=None:
                        child_list.append(node.right)
    
            father_list=child_list.copy()
            child_list=[]
        return depth


            
        
        
````
